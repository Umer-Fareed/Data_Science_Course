```python
import pandas as pd 
import seaborn as sn 
```

```python
df= pd.read_csv('bhp.csv')
df.head()
```

```python
df.price.describe()
```

```python
from matplotlib import pyplot as plt
plt.hist(df.price,
            bins= 20,
            rwidth= 0.9)
plt.yscale('log')
```

```python
df['zscore'] = ( df.price - df.price.mean() ) / df.price.std()
df.price.describe()
```

```python
df[df['zscore']>3]
```
