```python
import pandas as pd
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,3))
plt.plot(df_sales["Quarter"],df_sales["Fridge"], color='red',label = "fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"], color='red',label = "Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"], color='red',label = "Washing Machine")
plt.ylabel("q")
plt.xlabel("g")
plt.legend()
plt.show()

```
