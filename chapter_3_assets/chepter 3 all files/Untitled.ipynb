```python
def grouper(df , idx , col):
    if 80<=df[col].loc[idx] <= 90:
        return '80-90'
    elif 50<=df[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
g = df.groupby(lambda idx: grouper(df,idx,'temperature'))
g
```

```python
for key , d in g:
    print("\n")
    print("key",key)
    print("\n")
    print("data\n",d)
```

```python
def grouper(df , idx , col):
    if  1 <= df[col].loc[idx] <= 3.9:
           return 'poor'
    elif 4.0 <= df[col].loc[idx] <=7.9:
           return 'average'
    elif 8.0<= df[col].loc[idx] <=10.0:
           return "good"
    else:
        return "others"
```

```python
g = df3.groupby(lambda idx: grouper(df3 , idx, "imdb_rating"))
g
```

```python
for key, d in g:
    print("Group by Key: {}\n".format(key))
    print(d)
```
