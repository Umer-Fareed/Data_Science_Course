**LOAD LIBRARIES**

```python
import pandas as pd
```

**IMPORT DATA**

```python
df= pd.read_csv("carprices.csv")
df.head()
```

**IMPORT LABEL ENCODER**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
```

**FIT LABEL ENCODER**

```python
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
dfle
```

**DEFINE X AND Y**

```python
x = dfle.drop('Sell Price($)', axis= 'columns')
```

```python
y = dfle['Sell Price($)']
```

**FIT MODEL**

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x,y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict([[2,4000,3]])
```

**IMPORT DATA WITH MULTIPLE CATAGORICAL VERIABLES**

```python
df2 = pd.read_excel("Heart_Data.xlsx")
df2.head()
```

```python
dfle2 = df2
```

**CHANGE THE TYPE BY LABEL ENCODER**

```python

for col in dfle2.select_dtypes(include=['object']).columns:
    dfle2[col] = le.fit_transform(dfle2[col])
```

```python
dfle2.sample(10)
```

**DEFINE X AND Y**

```python
x2 = dfle2.drop('heart_attack', axis= 'columns')
```

```python
y2 = dfle2.heart_attack
```

```python
lm.fit(x2, y2)
```

```python
lm.coef_
```

```python
lm.intercept_
```

**IMPORT MODEL TO SAVE FILE**

```python
import pickle
```

```python
with open ('LabelEncoder','wb') as file:
    pickle.dump(lm,file)
```

```python
with open ('LabelEncoder', 'rb') as f:
    pickle.load(f)
```
