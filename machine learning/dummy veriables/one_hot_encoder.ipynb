```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
df_dumies = pd.concat([df, dummies], axis= 'columns')
df_dumies.head()
```

```python
df_dumies.drop(['town','west windsor'], axis='columns',inplace=True
                )
```

```python
df_dumies.head()
```

```python
x = df_dumies.drop('price', axis='columns')
```

```python
y = df_dumies.price
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x,y)
```

```python
lm.coef_
```

```python
lm.predict([[3000, 0,0]])
```

```python
lm.score
```

```python
lm.predict([[2800, 0, 1]])
```
