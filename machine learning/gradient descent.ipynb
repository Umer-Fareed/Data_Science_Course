```python


def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.001
    tolerance = 1e-9

    for i in range(iterations):
        yp = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y - yp)])
        md = -(2/n) * sum(x * (y - yp))
        bd = -(2/n) * sum(y - yp)
        m_prev, b_prev = m_curr, b_curr
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd
        if np.isclose([m_curr, b_curr], [m_prev, b_prev], atol=tolerance).all():
            break
        print("m {}, b {}, iteration {}, cost {}".format(m_curr, b_curr, i, cost))




# Call the gradient_descent function
gradient_descent(x, y)

```
