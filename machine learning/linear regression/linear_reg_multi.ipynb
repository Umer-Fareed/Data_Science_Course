```python
import pandas as pd
import numpy as np 
from sklearn import linear_model
import math 
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
#Filling missing values 
median_bedrooms = df.bedrooms.median()
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
df
```

```python
lm = linear_model.LinearRegression()
lm.fit(df[["area", "bedrooms", "age"]], df.price)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict([[3000, 3, 40]])
```

```python
df2 = pd.read_csv("hiring.csv")
df2
```

```python
from word2number import w2n
```

```python
df2.experience = df2.experience.fillna("zero")
df2
```

```python
df2.experience = df2.experience.apply(lambda x: w2n.word_to_num(x))
df2
```

```python
test_score_mean = math.floor(df2["test_score(out of 10)"].mean())
test_score_mean
```

```python
df2["test_score(out of 10)"] = df2["test_score(out of 10)"].fillna(test_score_mean)
df2
```

```python
lm = linear_model.LinearRegression()
lm.fit(df2[["experience","test_score(out of 10)","interview_score(out of 10)"]],
       df2["salary($)"])
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict([[9,10,8]])
```
